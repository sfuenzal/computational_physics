```python
%matplotlib inline
from scipy.special import gamma
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Item a)

```python
n = 600
m = 5
```

```python
def insane_functions(x):
    return np.array([2**(2**x), 2**(2**(x+1))])
```

```python
def functions(x):
    return np.array([np.sqrt(2)**math.log2(x), x**2, math.factorial(x), gamma(1+np.log2(x)), (3/2)**x, x**3, math.log2(x)**2, math.log2(math.factorial(x)), x**(1/np.log2(x)), math.log2(np.log2(x)), x*2**x, x**(math.log2(np.log(x))), math.log2(x), 1, 2**(math.log2(x)), math.log2(x)**math.log2(x), np.exp(x), 4**(math.log2(x)), math.factorial(x+1), np.sqrt(np.log2(x)), 2**np.sqrt(2*math.log2(x)), x, 2**x, x*math.log2(x)]) 
```

```python
name_of_functions = np.array(["$\sqrt{2}^{\log_{2}(n)}$", "$n^2$", "$n!$", "$\log_{2}(n)!$", "$(3/2)^n$", "$n^3$", "$\log^2_{2}(n)$", "$\log_{2}(n!)$", "$n^{1/\log_2(n)}$", "$\log_{2}\log_{2}(n)$", "$n2^n$", "$n^{\log_2\log_2(n)}$", "$\log_2(n)$", "$1$", "$2^{\log_{2}(n)}$", "$\log_{2}(n)^{\log_{2}(n)}$", "$e^n$", "$4^{\log_{2}(n)}$", "$(n+1)!$", "$\sqrt{\log_2(n)}$", "$2^{\sqrt{2\log_{2}(n)}}$", "$n$", "$2^n$", "$n\log_{2}(n)$"])
```

```python
name_of_iterated_functions = np.array(["$\log_{2}\left(\log^{*}_{2}(n)\right)$", "$2^{\log^{*}_{2}(n)}$", "$\log^{*}_{2}(n)$", "$\log^{*}_{2}\log_{2}(n)$"])
```

```python
name_of_insane_functions = np.array(["$2^{2^n}$", "$2^{2^{n+1}}$"])
```

```python
def iterated_function_3(x):
    if x <= 1:
        return 0
    if x > 1:
        return 1 + iterated_function_3(math.log2(x))    
```

```python
def iterated_function_1(x):
    return np.log2(iterated_function_3(x))
```

```python
def iterated_function_2(x):
    return 2**(iterated_function_3(x))
```

```python
def iterated_function_4(x):
    return iterated_function_3(math.log2(x))
```

```python
def iterated_functions(x):
    return np.array([iterated_function_1(x), iterated_function_2(x), iterated_function_3(x), iterated_function_4(x)])
```

```python
array_of_normal_function = np.array([str(functions(n)[i])+" "+name_of_functions[i] for i in range(len(functions(n)))])
```

```python
array_of_iterated_function = np.array([str(iterated_functions(n)[i])+" "+name_of_iterated_functions[i] for i in range(len(iterated_functions(n)))])
```

```python
array_of_insane_function = np.array([str(insane_functions(m)[i])+" "+name_of_insane_functions[i] for i in range(len(insane_functions(m)))])
```

```python
total_array = np.append(array_of_normal_function, array_of_iterated_function)
```
