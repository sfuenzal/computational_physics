```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
import time
```

```python
def multiplication_of_matrices(n, min_element, max_element):
    A = []
    for i in range(n):
        A.append(list(np.random.randint(min_element, high = max_element, size = n)))
    
    B = []
    for i in range(n):
        B.append(list(np.random.randint(min_element, high = max_element, size = n)))
    
    B_array = np.array(B)
    B_transpose_array = B_array.T
    B_transpose = B_transpose_array.tolist()
    
    C = []
    for i in range(len(A)):
        for j in range(len(B)):
            C.append([sum(np.multiply(A[i],B_transpose[j]))])
    
    C_array = np.array(C).reshape(n,n)
    
    #return print("C = AB =" + str(C_array), "A = " + str(A), "B = " + str(B), sep = "\n")
    return C_array
```

```python
multiplication_of_matrices(5,-5,5)
```

```python
maximum = 500
times = []
for i in range(1,maximum,10):
    t0 = time.time()
    multiplication_of_matrices(i,-5,5)
    t1 = time.time() - t0
    times.append(t1)
```

```python
times_array = np.array(times)
```

```python
lenghts_array = np.array([i for i in range(1,maximum,10)])
```

```python
def f(x):
    return (1e-4)*x**2.2
```

```python
times_array
```

```python
f(lenghts_array)
```

```python
plt.figure(figsize = (8,8))
plt.plot(lenghts_array, times_array, label = r'$\Theta(n^{2.2})$')
plt.plot(lenghts_array, 1*f(lenghts_array), label = r'$c2g(n)=c2n^{2.2}$')
plt.plot(lenghts_array, 0.3*f(lenghts_array), label = r'$c1g(n)=c1n^{2.2}$')
plt.xlabel(r'$n$', fontsize = 18)
plt.ylabel(r'$T(n)$', fontsize = 18)
plt.legend(loc = 'best', fontsize = 13)
plt.grid(b = True, which = 'major', linestyle = '-')
plt.savefig('problem-6-plot.pdf')
```
