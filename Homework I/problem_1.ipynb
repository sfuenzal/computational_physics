# Item d)

```python
def array_to_inversions(n):
    return list(np.array([i for i in range(n,0,-1)]))
```

```python
def half_array(n):
    return list(np.array([array_to_inversions(n)[:len(array_to_inversions(n))//2], array_to_inversions(n)[len(array_to_inversions(n))//2:]]))   
```

```python
parameter_1 = 2
parameter_2 = 3
```

```python
def inversion_code(n):
    i = parameter_1
    j = parameter_2
    while i < len(half_array(n)[0]) and j < len(half_array(n)[0]):
        if half_array(n)[0][i] > half_array(n)[0][j] and i < j:
            print(str(half_array(n)[0][i])+str(half_array(n)[0][j]))
            j = j + 1
    i = parameter_1
    j = parameter_2
    while i < len(half_array(n)[1]) and j < len(half_array(n)[1]) and i < j:
        if half_array(n)[1][i] > half_array(n)[1][j]:
            print(str(half_array(n)[1][i])+str(half_array(n)[1][j]))
            j = j + 1
    i = parameter_1
    j = parameter_2
    while i < len(half_array(n)[0]) and j < len(half_array(n)[1]) and i < j:
        if half_array(n)[0][i] > half_array(n)[1][j]:
            print(str(half_array(n)[0][i])+str(half_array(n)[1][j]))
            j = j + 1
```

```python
inversion_code(5)
```

```python
maximum = 400
times = []
for i in range(2,maximum,5):
    t0 = time.time()
    inversion_code(i)
    t1 = time.time() - t0
    times.append(t1)
```

```python
times_array = np.array(times)
```

```python
lenghts_array = np.array([i for i in range(2,maximum,5)])
```

```python
def f(x):
    return (1e-4)*x*np.log2(x)
```

```python
plt.figure(figsize = (8,8))
plt.plot(lenghts_array, times_array, label = r'$\Theta(n\log_{2}(n))$')
plt.plot(lenghts_array, 0.9*f(lenghts_array), label = r'$c1g(n)=c1n\log_{2}(n)$')
plt.plot(lenghts_array, 5*f(lenghts_array), label = r'$c2g(n)=c2n\log_{2}(n)$')
plt.xlabel(r'$n$', fontsize = 18)
plt.ylabel(r'$T(n)$', fontsize = 18)
plt.legend(loc = 'best', fontsize = 13)
plt.grid(b = True, which = 'major', linestyle = '-')
plt.savefig('problem-1-d-plot.pdf')
```
