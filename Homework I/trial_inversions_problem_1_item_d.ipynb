```python
%matplotlib inline
import numpy as np
import time
import matplotlib.pyplot as plt
import random
```

```python
parameter_1 = 3 #vary the parameter1 from 0 to any number
parameter_2 = 4 #vary the parameter2 from 0 to any number
#Beware: Always check that parameter2-paramter1 = 1
```

```python
def array(n):
    return list(np.array([i for i in range(n,0,-1)]))
```

```python
def inversions(n):
    i = parameter_1
    j = parameter_2
    while i < len(array(n)) and j < len(array(n)):
        if array(n)[i] > array(n)[j]:
            #print(str(array(n)[i]) + str(array(n)[j]))
            j = j + 1
```

```python
inversions(7)
```

```python
maximum = 1000
times = []
for i in range(2,maximum,5):
    t0 = time.time()
    inversions(i)
    t1 = time.time() - t0
    times.append(t1)
```

```python
times_array = np.array(times)
```

```python
lenghts_array = np.array([i for i in range(2,maximum,5)])
```

```python
def f(x):
    return (1e-5)*x*np.log2(x)
```

```python
plt.figure(figsize = (8,8))
plt.plot(lenghts_array, times_array, label = r'$\Theta(n\log_{2}(n))$')
plt.plot(lenghts_array, 0.9*f(lenghts_array), label = r'$c1g(n)=c1n\log_{2}(n)$')
plt.plot(lenghts_array, 12.5*f(lenghts_array), label = r'$c2g(n)=c2n\log_{2}(n)$')
plt.xlabel(r'$n$', fontsize = 18)
plt.ylabel(r'$T(n)$', fontsize = 18)
plt.legend(loc = 'best', fontsize = 13)
plt.grid(b = True, which = 'major', linestyle = '-')
plt.savefig('problem-1-d-plot-trial.pdf')
```
