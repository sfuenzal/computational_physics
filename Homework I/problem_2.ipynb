# Problem 2

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Item a)

```python
imput_size_1 = np.array([i for i in range(1,100,1)])
```

```python
def case_1(n,k,epsilon,constant):
    return np.array([np.log2(n)**k, constant*n**epsilon])
```

```python
array_with_names_of_functions_1 = np.array(["$f(n)=\log^{k}_{2}(n)$", "$c2g(n)=c2n^{\epsilon}$"])
```

```python
plt.figure(figsize = (10,8))
for i in range(len(case_1(imput_size_1,2,3,0.001))):
    plt.plot(imput_size_1, case_1(imput_size_1,2,3,0.001)[i], label = array_with_names_of_functions_1[i])
    plt.legend(loc = 'best',fontsize=15)
    plt.grid(b = True, which = 'major', linestyle = '-')
    plt.xlabel(r'$n$', fontsize = 18)
    plt.savefig('problem-2-a-plot.pdf')
```

# Item b)

```python
imput_size_2 = np.linspace(0,20,100)
```

```python
def case_2(n,k,c,constant):
    return np.array([n**k, constant*c**n])
```

```python
array_with_names_of_functions_2 = np.array(["$f(n)=n^k$","$c1g(n)=c1c^n$"])
```

```python
plt.figure(figsize = (10,8))
for i in range(len(case_2(imput_size_2,2,3,1e-2))):
    plt.plot(imput_size_2, case_2(imput_size_2,2,2,1e-2)[i], label = array_with_names_of_functions_2[i])
    plt.legend(loc = 'best',fontsize=15)
    plt.xlim(0,20)
    plt.ylim(0,2000)
    plt.grid(b = True, which = 'major', linestyle = '-')
    plt.xlabel(r'$n$', fontsize = 18)
    plt.savefig('problem-2-b-plot.pdf')
```

# Item e)

```python
imput_size_5 = np.linspace(1,100,1000)
```

```python
def case_5(n,c,constant):
    return np.array([n**(np.log2(c)), constant*c**(np.log2(n)), constant*c**(np.log2(n))])
```

```python
array_with_names_of_functions_5 = np.array(["$f(n)=n^{\log_{2}(c)}$","$c1g(n)=c1c^{\log_{2}(n)}$","$c2g(n)=c2c^{\log_{2}(n)}$"])
```

```python
array_with_constants_1 = np.array([1, 0.5, 2])
```

```python
plt.figure(figsize = (10,8))
for i in range(len(array_with_constants_1)):
    plt.plot(imput_size_5, case_5(imput_size_5,3,array_with_constants_1[i])[i], label = array_with_names_of_functions_5[i])
    plt.legend(loc = 'best',fontsize=15)
    plt.grid(b = True, which = 'major', linestyle = '-')
    plt.xlabel(r'$n$', fontsize = 18)
    plt.savefig('problem-2-e-plot.pdf')
```

# Item f) 

```python
imput_size_6 = np.linspace(1,100,1000)
```

```python
def case_6(n,constant):
    return np.array([np.log2(np.sqrt(2*np.pi*n)*n**n*np.exp(-n)), constant*np.log2(n**n), constant*np.log2(n**n)])
```

```python
array_with_names_of_functions_6 = np.array(["$f(n)=\log_{2}(n!)$", "$c1g(n)=c1\log_{2}(n^n)$", "$c2g(n)=c2\log_{2}(n^n)$"])
```

```python
array_with_constants_2 = np.array([1, 1e-1, 2])
```

```python
plt.figure(figsize = (10,8))
for i in range(len(array_with_constants_2)):
    plt.plot(imput_size_6, case_6(imput_size_6,array_with_constants_2[i])[i], label = array_with_names_of_functions_6[i])
    plt.legend(loc = 'best',fontsize=15)
    plt.grid(b = True, which = 'major', linestyle = '-')
    plt.xlabel(r'$n$', fontsize = 18)
    plt.savefig('problem-2-f-plot.pdf')
```
