```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# b)

```python
f1 = np.genfromtxt("times_file_1.txt")
```

```python
print(f1)
```

```python
aux1 = np.empty(len(f1))
for i in range(len(f1[0])):
    np.append(aux1,f1[i][len(f1[0])-1])
```

```python
lenght_array_1 = np.sort(np.append(np.array([i*i for i in range(2,len(aux1)+1)]),2))
```

```python
def f(x):
    return 1e-7*x**(np.log2(7))
```

```python
plt.figure(figsize = (10,8))
plt.plot(lenght_array_1, aux1, label = r'$\Theta(n^{\log(7)})$')
plt.plot(lenght_array_1, 0.1*f(lenght_array_1), label = r'$c1g(n)=c1n^{\log(7)}$')
plt.plot(lenght_array_1, 8*f(lenght_array_1), label = r'$c2g(n)=c2n^{\log(7)}$')
plt.xlabel(r'$n$', fontsize = 18)
plt.ylabel(r'$T(n)$', fontsize = 18)
plt.legend(loc = 'best', fontsize = 13)
plt.grid(b = True, which = 'major', linestyle = '-')
plt.savefig('plot_problem_2_item_b.pdf')
```
